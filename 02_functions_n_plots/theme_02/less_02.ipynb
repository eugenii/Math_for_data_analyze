```python
import numpy as np
import matplotlib.pyplot as plt

def deposit(months):
	return 100000 * (12.07 / 12) ** months


months = np.arange(61) # пять лет = 60 месяцев, и плюс нулевой день открытия депозита
depo = deposit(months)

print(deposit(12))
plt.plot(months, depo)
plt.show()
```

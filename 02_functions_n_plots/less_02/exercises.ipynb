```python
import numpy as np
```

```python
# sample

# Создаём линейную функцию.
def f(x):
    return 3 * x + 4

# Применяем функцию ко всем целым числам в промежутке [-5, 5).
print(f(np.arange(-5, 5)))

# Это также можно посчитать явно, без использования функции Python:
x = np.arange(-5, 5)
print(3 * x + 4)
```

```python
bones = np.array([28, 32, 36, 42])
# Создайте функцию women_height для вычисления роста женщин по длине плечевой кости.
# Создайте функцию men_height для вычисления роста мужчин по длине плечевой кости.
def women_height(bones):
    return 2.75 * bones + 71.48

def men_height(bones):
    return 2.89 * bones + 70.64

print(women_height(bones))
print(men_height(bones))
```

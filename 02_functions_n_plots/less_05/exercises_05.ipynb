В данных стоматологической клиники показано, как в среднем с возрастом меняется доля зубов (в процентах), поражённых кариесом. В первом столбце — средняя доля поражённых зубов в процентах, во втором — возраст.
Найдите коэффициенты полинома степени 
3
,
 который описывает зависимость доли поражённых зубов от возраста.
Визуализируйте данные и полином. При визуализации значений полинома используйте значения возраста из множества 
[
20

[20, 70) лет с шагом в полгода.
Вычислите значение полинома для возраста 
35
35 лет, сохраните в переменную caries_35 и выведите на экран.

```python
# Данные
caries_data = np.array([[ 4.4, 25. ],
						[12.4, 35. ],
						[15.5, 45. ],
						[14.8, 55. ],
						[12.4, 65. ]])
part = caries_data[:, 0]
years = caries_data[:, 1]
coefs = np.polyfit(part, years, 3)
f = np.poly1d(coefs)
x = np.arange(20, 70, 0.5)
y = f(x)
plt.plot(x, y)
caries_35 = f(35)
print(caries_35)
```

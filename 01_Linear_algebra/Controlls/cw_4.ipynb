```python
# import

import numpy as np
```

```python
# 1

a = np.array([
    [-8, 3, 9, 6],
    [0, 6, 0, 3],
    [1, 2, 78, 4.7],
    [2, 315, 0.6, 7]
])

b = np.linalg.inv(a)
# Допишите код здесь.
print("Матрица:")
print(a)
print("Обратная матрица:")
print(b)
print("Проверка единичной матрицей:")
print(np.dot(a, b).round())
```

```python
# 2

# Данные:

w = np.array([
    [-7, 4],
    [5, 8]
])

q = np.array([
    [9, -11, 56],
    [3, -6, -8],
    [2, 17, 4]
])
w_det = np.linalg.det(w)
q_det = np.linalg.det(q)

# Допишите код здесь.
print("Матрица:")
print(w)
print("Определитель:")
print(w_det)
print("Матрица:")
print(q)
print("Определитель:")
print(q_det)
```

```python
# 3

P = np.array([
    [-5, 1, 3, 6],
    [-3, 2, -1, 8],
    [4, -12, 9, 0],
    [0, 3, 6, 2]
])

Y = np.array([3, 11, -6, 5])
Y_column = Y.reshape(4, 1)
F = np.array([8, -2, 1, 3])
F_column = F.reshape(4, 1)

# Допишите код здесь.
X_preimage = np.linalg.solve(P, Y_column)
F_image = np.dot(P, F_column)
print("Прообраз точки Y:")
print (X_preimage)
print("Проверка:")
print(np.dot(P, X_preimage))
print("Образ точки F:")
print(F_image)
print("Проверка:")
print(np.linalg.solve(P, F_image))
```

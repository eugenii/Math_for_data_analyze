```python
# Данные:
A = np.array([2, -12, 8.5])
B = np.array([0.7, 21.2, 0.8])
C = np.array([5, -0.6, -4])
D = np.array([1, -8.2, 5])
# 2, -12, 8.5
# 0.7, 21.2, 0.8
# 5, -0.6, -4
# 1, -8.2, 5
BA = A - B
BC = C - B

cos_angle  = np.dot(BA, BC) / (np.linalg.norm(BA) * np.linalg.norm(BC))
angle = np.arccos(cos_angle)    # допишите код для вычисления в радианах здесь
print("Косинус угла:", cos_angle)
print("Сам угол:", np.rad2deg(angle))
```

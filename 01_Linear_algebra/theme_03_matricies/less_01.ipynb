Запись матрицы в коде

```python
import numpy as np

m = np.array([
      [2, 17, 43],
      [36, 8, 4],
      [10, -11, 6],
      [3, 7, 15]
    ])

k = np.array([
    [1],
    [14],
    [6],
    [9]
    ])
print("Матрица:")  # Вывод подписи на экран.
print(m)
print()  # Пустая строка для разделения.
print("Вектор-столбец:")  # Вывод подписи на экран.
print(k) 
```

```python
# Транспонированная матрица
m = np.array([
    [2, 17, 43],
    [36, 8, 4],
    [10, -11, 6],
    [3, 7, 15]
    ])
print("Матрица:")
print(m)  # Вывод исходной матрицы.
print()  # Пустая строка для разделения.
m = m.T  # Теперь m - это транспонированная матрица.
print("Транспонированная матрица:")
print(m)  # Вывод новой, транспонированной матрицы. 
```

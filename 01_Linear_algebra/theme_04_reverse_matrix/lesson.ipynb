```python
import numpy as np

a = np.array([
    [0.55, 0.4, 1.2, 6],
    [-12, -9, 0.5, 1],
    [8.8, -8, -4, 2.9],
    [1.3, 7, -0.94, 2]
])

# Допишите код здесь.

print("Матрица:")
print(a)
print("Обратная матрица:")
# Пожалуйста, сохраните обратную матрицу в переменной b
b = np.linalg.inv(a)
print(b)
print("Проверка единичной матрицей:")
print((a @ b).round())
```

```python
import numpy as np

a = np.array([
    [325, 81],
    [260, 60]
])
b = np.array([2599, 2060])
o = np.linalg.inv(a) # Вычисление обратной матрицы.
b_column = b.reshape(2, 1)
result = o @ b_column # Скалярное произведение.

print(result)
```

```python
#  Определитель

import numpy as np

# Данные:
a = np.array([
    [-0.8, 24],
    [1, 3.7]
])

# Допишите код здесь.
print("Матрица:")
print(a)
print("Определитель:")
det = np.linalg.det(a)
print(det)
```

```python
import numpy as np

a2 = np.array([-3, -2])    # Вектор а1
a1 = np.array([6, -3])     # Вектор а2
a = np.array([a1, a2])     # Матрица
det = np.linalg.det(a)     # Определитель матрицы

print("Матрица:")
print(a)
print("Определитель:")
print(det)
```

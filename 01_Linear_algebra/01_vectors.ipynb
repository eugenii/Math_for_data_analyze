```python
### Создание вектора в numpy

import numpy as np
```

```python
# import numpy as np

vec_a = np.array([1, 2, 3])

# нулевой вектор с размерностью 5
vec_b = np.zeros(5)
length_a = len(vec_a) # длина вектора
length_b = len(vec_b)
print(length_a, length_b)
```

```python
# Сложение векторов в координатах (многомерное)
# Пример со стоимостью товаров по месяцам

# import numpy as np

january = np.array([648, 371, 338.4, 512, 0, 437, 11])
february = np.array([112, 380, 516, 294, 108, 380, 65])
march = np.array([506, 0, 412, 98, 304, 811, 56])

c =  january + february + march

print(c)
```
